```python
# nltk.download("vader_lexicon")
import pandas as pd
data = pd.read_csv('output/preprocessed_data6.csv')
data.head()
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sentiments = SentimentIntensityAnalyzer()
# data["Positive"] = [sentiments.polarity_scores(i)["pos"] for i in data["case_fold"]]
# data["Negative"] = [sentiments.polarity_scores(i)["neg"] for i in data["case_fold"]]
# data["Neutral"] = [sentiments.polarity_scores(i)["neu"] for i in data["case_fold"]]
data["Compound"] = [sentiments.polarity_scores(i)["compound"] for i in data["case_fold"]]
data.head()
```

```python
score = data["Compound"].values
sentiment = []
for i in score:
  if i >= 0.05:
    sentiment.append('Positive')
  elif i <= -0.05:
    sentiment.append('Negative')
  else:
    sentiment.append('Neutral')
data["Sentiment"] = sentiment
data.head(10)
```

```python
data.to_csv('output/vader_labelled7.csv', index=False)
```
