```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

fig, ax = plt.subplots(figsize=(12, 10))

# Define the boxes and arrows
flowchart = [
    {"text": "1. Pemilihan Fitur Lexical", "xy": (0.5, 0.9)},
    {"text": "2. Pengumpulan Penilaian Intensitas Sentimen", "xy": (0.5, 0.75)},
    {"text": "3. Validasi Melalui Pendekatan Human-Centric", "xy": (0.5, 0.6)},
    {"text": "4. Proses Rating Menggunakan Antarmuka Khusus", "xy": (0.5, 0.45)},
    {"text": "5. Contoh Rating dan Skor Valensi", "xy": (0.5, 0.3)},
    {"text": "6. Validasi dan Penyesuaian Akhir", "xy": (0.5, 0.15)},
    {"text": "7. Penggunaan Skor Valensi dalam Analisis Sentimen", "xy": (0.5, 0.0)}
]

# Draw the boxes
for step in flowchart:
    ax.text(step["xy"][0], step["xy"][1], step["text"],
            ha="center", va="center", fontsize=12,
            bbox=dict(boxstyle="round,pad=0.3", edgecolor="black", facecolor="#c0c0c0"))

# Draw the arrows
for i in range(len(flowchart) - 1):
    ax.annotate("", xy=flowchart[i+1]["xy"], xytext=flowchart[i]["xy"],
                arrowprops=dict(arrowstyle="->", lw=2))

# Set limits and hide axes
ax.set_xlim(0, 1)
ax.set_ylim(-0.1, 1)
ax.axis('off')

plt.show()
```
