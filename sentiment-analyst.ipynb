```python
def clean_tweet(text):
  text = re.sub(r'@[A-Za-z0-9_]+', '', text)
  text = re.sub(r'#\w+', '', text)
  text = re.sub(r'https?://\S+', '', text)
  
  text = re.sub(r'[^A-Za-z0-9 ]', '', text)
  text = re.sub(r'\s+', ' ', text).strip()
  
  return text

df['second_cleaned'] = df['stemmed'].apply(clean_tweet)
df
```

```python
df = df.drop_duplicates(subset=['second_cleaned', 'full_text'], keep='first')
translator = Translator(service_urls=['translate.googleapis.com']) 

def google_translate(word, src='id', dest='en'):
    translator = Translator()
    translation = translator.translate(word, src=src, dest=dest)
    return translation.text
  
def clean_tweet2(text):
    text = re.sub('https?://\S+|www\.\S+', '', text)
    text = re.sub(r'\s+', ' ', text, flags=re.I)
    text = re.sub('\[.*?\]', '', text)
    text = re.sub('\n', '', text)
    text = re.sub('\w*\d\w*', '', text)
    text = re.sub('<.*?>+', '', text)
    return text
```

```python
df['label'] = df['second_cleaned'].apply(lambda x: google_translate(x, src='id', dest='en'))
df.to_csv("translated_data.csv")
```
