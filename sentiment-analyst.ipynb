```python
def clean_tweet(text):
  text = re.sub(r'@[A-Za-z0-9_]+', '', text)
  text = re.sub(r'#\w+', '', text)
  text = re.sub(r'https?://\S+', '', text)
  
  text = re.sub(r'[^A-Za-z0-9 ]', '', text)
  text = re.sub(r'\s+', ' ', text).strip()
  
  return text

df['second_cleaned'] = df['lemmatize'].apply(clean_tweet)
df
```

```python
df = df.drop_duplicates(subset=['second_cleaned', 'full_text'], keep='first')

def translate_text( target: str, text: str) -> dict:
  from google.cloud import translate_v2 as translate
  translate_client = translate.Client()
  result = translate_client.translate(text, target_language=target)
  return result

def clean_tweet2(text):
    text = re.sub('https?://\S+|www\.\S+', '', text)
    text = re.sub(r'\s+', ' ', text, flags=re.I)
    text = re.sub('\[.*?\]', '', text)
    text = re.sub('\n', '', text)
    text = re.sub('\w*\d\w*', '', text)
    text = re.sub('<.*?>+', '', text)
    return text
```

```python
import json
from html import unescape
import string

def clean_text(text):
    # Hapus mention (@username) menggunakan regex
    text = re.sub(r'@\w+', '', text)
    # Hapus hashtag (#hashtag) menggunakan rege
    text = re.sub(r'#\w+', '', text)
    # Hapus URL menggunakan regex
    text = re.sub(r'http\S+|www\S+|https\S+', '', text)
    # Hapus emoji menggunakan regex
    text = re.sub(r'[\U0001F600-\U0001F64F\U0001F300-\U0001F5FF\U0001F680-\U0001F6FF\U0001F700-\U0001F77F]+', '', text)
    # Hapus semua tanda baca menggunakan regex
    # text = re.sub(r'[^\w\s]', '', text)
    text = text.translate(str.maketrans('', '', string.punctuation))
    return text

def extract_text_value(text):
    # Regular expression to find and extract value after 'text='
    pattern = r'text=([^,]*)'
    # Search the text using the regular expression
    match = re.search(pattern, text)
    # If a match is found, return the value
    if match:
        # Strip leading and trailing whitespaces and quotes
        return match.group(1).strip().strip("'").strip('"')
    else:
        # Return None if no match is found
        return None

df['translated'] = df['translated'].apply(extract_text_value)
df['translated'] = df['translated'].apply(clean_text)
df
```

```python
def remove_b_prefix(text):
    # Check if the text is a byte string and starts with 'b' prefix
    if isinstance(text, bytes) and text.startswith(b'b'):
        # Remove the 'b' prefix and decode the bytes object to a regular string
        return text[1:].decode('utf-8')
    elif isinstance(text, str):
        # If the input is a string, just return it as is
        return text
    else:
        # If it's neither a string nor bytes, raise an error
        raise ValueError("Input must be a str or bytes object.")

df['translated'] = df['translated'].apply(remove_b_prefix)
df.head(5)
```
